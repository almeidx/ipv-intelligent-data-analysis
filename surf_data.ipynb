```python
import pandas as pd
df = pd.read_excel('gonogowindsurfing_clean.xlsx')
```

```python
pd.isnull(df)
```

```python
df.loc[pd.isnull(df.windSpeed)]
```

```python
df.loc[df.windSpeed.isnull(), 'windSpeed'] = df.windSpeed.mean()
```

```python
df.loc[df.temperature.isnull(), 'temperature'] = df.temperature.mean()
```

```python
# df.drop(df[df.age < 0].index, inplace=True)

df.loc[df.age <= 0, 'age'] = df.age[df.age > 0].mean()
```

```python
df['age'] = df['age'].astype(int)
```
